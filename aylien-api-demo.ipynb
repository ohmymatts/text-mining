```python
from aylienapiclient import textapi
import csv, io
```

```python
#criando a conexão com a API
cliente = textapi.Client("c0854be6", "387ec926d87f2c61f37d9ee52179ee85")

with io.open("avanade.csv", "w", encoding="utf8", newline="") as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(["Tweet", "Sentiment"])
    with io.open("avanade.txt", "r", encoding="utf8") as f:
        for tweet in f.readlines():
            
            tweet = tweet.strip()
            
            if len(tweet) == 0:
                print("não computado")
                continue
        
            print(tweet)
        
        #chamando a api
            sentiment = cliente.Sentiment({'text': tweet})
        
            csv_writer.writerow([sentiment['text'], sentiment['polarity']])
```
